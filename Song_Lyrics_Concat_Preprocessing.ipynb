```python
import numpy as np 
import pandas as pd
import spacy
```

```python
def stripBrackets(text):
    text = text.strip('[')
    text = text.strip(']')
    text = text.strip('"')
    text = text.strip('\'')
    text = text.strip()
    return text
```

```python
def loadData():
    class_labels = ['happy', 'sad', 'hate', 'romantic', 'peaceful', 'motivational', 'funny']
    dfs = []
    for label in class_labels:
        df = pd.read_csv('{}_lyrics.csv'.format(label))
        df = df[['Artist', 'Song', 'Lyrics', 'Class']]
        df['Lyrics'] = df['Lyrics'].apply(stripBrackets)
        dfs.append(df)
    for df in dfs:
        display(df)
        
    return pd.concat(dfs, axis=0).reset_index(drop=True)
```

```python
df = loadData()
```

```python
import spacy
nlp = spacy.load("en_core_web_md")
```

```python
def preprocess_text(string):
  # Replacing the unicode character
  string = string.replace('\u200e'," ")
  # Seperating the string into lines
  lines = string.split("\n")
  
  # Take max 100 lines
  lines = lines[:100]

  # Initialzing a set to avoid duplicate lines
  uniq_lines = set()
  new_lines = []
  for line in lines: 
    # Removes the leading and trailing white spaces
    new_line = line.strip()
    # Converts text to lower case
    new_line = new_line.lower()
    # Skip the line if it's empty
    if not new_line:
      continue
    if 'lyrics' in new_line.lower():
      continue
    if new_line.startswith('[') and new_line.endswith(']'):
      continue
    if new_line not in uniq_lines:
      new_lines.append(new_line)
      uniq_lines.add(new_line)
  
  # Take max 50 unique lines
  new_lines = new_lines[:50]

  result = []
  for line in new_lines:
    tokens = []
    for token in nlp(line):
      # Removes punctuation, currency, digits, numbers, oov words, spaces, stop words and proper nouns
      if not token.is_punct and not token.is_currency and not token.is_digit and not token.is_oov and not token.is_space and not token.is_stop and not token.like_num and not token.pos_ == "PROPN":
        # Add the lemmatized version of the token
        if token.lemma_ in ['o', 'oh', 'ohh', 'na', 'yeah']:continue
        tokens.append(token.lemma_)
    result.append(" ".join(tokens))
      
  return " ".join(result)
```

```python
df['Lyrics'] = df['Lyrics'].apply(preprocess_text)
```

```python
# Remove songs with invalid lyrics
df = df[~df['Lyrics'].isna()].reset_index(drop=True)

# Remove songs with invalid lyrics
df = df[df['Lyrics'] != ''].reset_index(drop=True)
```
