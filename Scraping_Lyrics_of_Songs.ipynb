```python
def get_artist_and_song(title):
  artist_name,song_name = title.split("-",1)
  index = song_name.find("(")
  if index!=-1:
    song_name = song_name[:index]
  return artist_name,song_name
```

```python
artists = []
songs = []
count = 0
for song in sad['Song Name']:
  try:
    artist,song = get_artist_and_song(song)
  except:
    print("_____________________________")
    print(song)
    print("_____________________________")
    continue
  print(artist,song)
  count+=1
  artists.append(artist)
  songs.append(song)
```

```python
print(count)
```

```python
sad_data = list(zip(artists,songs))
```

```python
sad_data = [list(x) for x in sad_data]
```

```python
sad_preprocessed_data = pd.DataFrame(sad_data,columns = ['Artist','Song Name'])
sad_preprocessed_data.head()
```

```python
sad_preprocessed_data.to_csv("sad_preprocessed_data.csv",index=False)
```
